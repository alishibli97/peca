# Data Preprocessing

```python
df.dtypes
```

**We see that "Country" and "Status" columns are object columns (String values), we can convert them to integers using LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder

df_transformed = df.copy()

le_country = LabelEncoder()
le_status = LabelEncoder()

le_country.fit(df.Country)
df_transformed['Country'] = le_country.transform(df_transformed.Country)

le_status.fit(df.Status)
df_transformed['Status'] = le_status.transform(df_transformed.Status)
```

**to revert the encoding we can do: df['Country'] = le_country.inverse_transform(sub_df['Country'])**

```python
df_transformed.dtypes
```

```python
df_transformed.head()
```

```python
df_transformed.describe()
```

**We notice that the Country and Year columns have multiple discrete values. It is a good idea at this point to transform this into a one-hot-encoding vector. The reason for this is for better representation. The points will become equally far distant from each other. So distance between Country 1 and Country 15 is the same as the distance between Country 1 and Country 2.**

```python
from sklearn.preprocessing import LabelBinarizer

country_encoder = LabelBinarizer()
transformed = country_encoder.fit_transform(df_transformed['Country'])
ohe_df = pd.DataFrame(transformed, columns = ["Country_%d" % (i + 1) for i in range(len(set(df_transformed.Country)))])
df_expanded = pd.concat([df_transformed, ohe_df], axis=1).drop(['Country'], axis=1)

year_encoder = LabelBinarizer()
transformed = year_encoder.fit_transform(df['Year'])
ohe_df = pd.DataFrame(transformed, columns = ["Year_%d" % (i + 1) for i in range(len(set(df_transformed.Year)))])
df_expanded = pd.concat([df_expanded, ohe_df], axis=1).drop(['Year'], axis=1)

df_expanded
```

```python
df_expanded.describe()
```

**Finally, we would want to scale our datapoints to the same range. This is essential. The vectors inputted to the machine learning models are mathematical vectors. The code understands numbers, and numbers need to be of similar ranges, otherwise the model will favour high numbers on low numbers even though we didn't meet that. Example: some columns come by default with high values (Measles) and some with low values (Alcohol). Its our job as Data Scientist to solve these issues and curate our data.**

**We can scale the data by specifying the columns we want to scale with the StandardScaler function.**

```python
from sklearn.preprocessing import StandardScaler

columns = list(df_expanded.columns)
columns.remove("Life expectancy")
columns_to_normalize = columns

scaler = StandardScaler()
df_expanded[columns_to_normalize] = scaler.fit_transform(df_expanded[columns_to_normalize])
```

**Divide data into labels X and targets y**

```python
from sklearn.model_selection import train_test_split

X = df_expanded.drop('Life expectancy',axis=1)
y = df_expanded['Life expectancy']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

1. LinearRegression

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train,y_train)
reg.score(X_test, y_test)
```

2. Support Vector Regressor

```python
from sklearn.svm import SVR

svr = SVR().fit(X_train,y_train)
svr.score(X_test, y_test)
```

3. Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor().fit(X_train,y_train)
regressor.score(X_test, y_test)
```

**From the docs the score is defined as:**

Return the coefficient of determination of the prediction.

The coefficient of determination R2 is defined as (1 - u/v), where u is the residual sum of squares ((y_true - y_pred)^2).sum() and  v is the total sum of squares ((y_true - y_true.mean())^2).sum(). The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). A constant model that always predicts the expected value of y, disregarding the input features, would get a R2 score of 0.0.

**Pick the model with the highest score and do predictions on the data**

```python
y_train_pred = regressor.predict(X_train)
y_test_pred = regressor.predict(X_test)
```

**Check the mean squared error between predicted and actual values (of life expectancy)**

```python
from sklearn.metrics import mean_squared_error as mse

print(f"Train error = {mse(y_train,y_train_pred)}")
print(f"Test error = {mse(y_test,y_test_pred)}")
```

Regularization II: Ridge
Lasso is great for feature selection, but when building regression models, Ridge regression should be your first choice.

Recall that lasso performs regularization by adding to the loss function a penalty term of the absolute value of each coefficient multiplied by some alpha. This is also known as  regularization because the regularization term is the  norm of the coefficients. This is not the only way to regularize, however.

If instead you took the sum of the squared values of the coefficients multiplied by some alpha - like in Ridge regression - you would be computing the  norm. In this exercise, you will practice fitting ridge regression models over a range of different alphas, and plot cross-validated  scores for each, using this function that we have defined for you, which plots the  score as well as standard error for each alpha:

def display_plot(cv_scores, cv_scores_std):

    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(alpha_space, cv_scores)

    std_error = cv_scores_std / np.sqrt(10)

    ax.fill_between(alpha_space, cv_scores + std_error, cv_scores - std_error, alpha=0.2)
    ax.set_ylabel('CV Score +/- Std Error')
    ax.set_xlabel('Alpha')
    ax.axhline(np.max(cv_scores), linestyle='--', color='.5')
    ax.set_xlim([alpha_space[0], alpha_space[-1]])
    ax.set_xscale('log')
    plt.show()
    
    
Don't worry about the specifics of the above function works. The motivation behind this exercise is for you to see how the  score varies with different alphas, and to understand the importance of selecting the right value for alpha. You'll learn how to tune alpha in the next chapter.
